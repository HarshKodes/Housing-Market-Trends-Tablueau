```python
import pandas as pd

# Step 1: Load the dataset
# Replace 'housing_data.csv' with your actual dataset file path
df = pd.read_csv("Housing.csv")

# Step 2: Basic dataset overview
print("🔹 Dataset Overview")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\nColumn Names:", df.columns.tolist())

# Step 3: Info about columns (types, nulls)
print("\n🔹 Column Info")
print(df.info())

# Step 4: Quick statistics for numerical columns
print("\n🔹 Numerical Summary")
print(df.describe())

# Step 5: Quick look at categorical columns
print("\n🔹 Categorical Value Counts")
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts().head(10))  # top 10 categories

# Step 6: Missing values check
print("\n🔹 Missing Values")
print(df.isnull().sum())
```
